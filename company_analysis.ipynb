```python
# team distribution
team_count=df["Team"].value_counts()
```

```python
team_percentage=(team_count / len(df))*100
```

```python
print("\n Team distribution ")
print(team_count)
```

```python
print("\n Team percentage ")
print(team_percentage)
```

```python
# team_percentage visualization
plt.figure(figsize=(8, 6)) 
team_percentage.plot(kind='bar',color='green')
plt.title("Percentage of employee distribution")
plt.xlabel("Team")
plt.ylabel("team_employee_percentage")
plt.show()
```

```python
# seggregate employees based on their position within company
positions=df.groupby("Position")
```

```python
for position, group in positions:
    print(f"\nEmployees in Position: {position}")
    print(group)
```

```python
# predominant age group
age_count=df["Age"].value_counts()
print(age_count)
```

```python
predominant_age=age_count.index[0]
print(f"predominant age group is {predominant_age}")
```

```python
# visualization of age group
plt.figure(figsize=(8,6))
```

```python
age_count.plot(kind='bar' ,color="blue")
plt.title("Age group distribution")
plt.xlabel("Age of employees")
plt.ylabel("No.of employees")
plt.show()
```

```python
# Highest Salary Expenditure (Team and Position)
salary_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum()
```

```python
highest_expenditure = salary_by_team_position.idxmax()
print(f"\nTeam and Position with Highest Salary Expenditure: {highest_expenditure}")
```

```python
# visualization of heights salary expenditure
salary_by_team_position_visual=salary_by_team_position.unstack()
plt.figure(figsize=(8,6))
snc.heatmap(salary_by_team_position_visual,annot=True , cmap="YlGnBu")
plt.title("Salary Expenditure (Team and Position)")
plt.xlabel("position")
plt.ylabel("Team")
plt.show()            
```

```python
# correlation between age and salary
correlation=df["Age"].corr(df["Salary"])
print(f"correlation between age and salary is {correlation}")
```

```python
# visualization of correlation
plt.figure(figsize=(8,6))
plt.scatter(df["Age"],df["Salary"], alpha=0.7,color='coral')
plt.title("correlation between age and salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.tight_layout()
plt.show()
```
